```python
message = input("Enter message: ").lower()

spam_words = ["lottery", "free", "winner"]

for word in spam_words:
    if word in message:
        print("⚠️ Spam Detected")
        break
else:
    print("✅ Safe Message")
```

```python
import pandas as pd

# Create a dummy DataFrame to save as students.csv
data = {
    "Student_Name": ["nisa", "dhaya", "tom"],
    "Age": [20, 22, 21],
    "Grade": ["A", "B", "A+"]
}
dummy_df = pd.DataFrame(data)

# Save the DataFrame to a CSV file named students.csv
dummy_df.to_csv("students.csv", index=False)

# Now, read the students.csv file
df = pd.read_csv("students.csv")

print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Name": ["nisa", "dhaya", "tom"],
    "Mark": [90, 95, 85]
}

df = pd.DataFrame(data)

plt.bar(df["Name"], df["Mark"])

plt.xlabel("Students")
plt.ylabel("Marks")

plt.title("Student Performance")

plt.show()
```
